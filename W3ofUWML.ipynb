#build the word count vector

```python
products['word_count']=graphlab.text_analytics.count_words(products['review'])
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 
                  'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def awesome_count(line):
    if ('awesome' in line):
        return int(line['awesome'])
    else:
        return 0
products['awesome'] = products['word_count'].apply(awesome_count)
def great_count(line):
    if 'great' in line:
        return line['great']
    else:
        return 0
products['great'] = products['word_count'].apply(great_count)
def fantastic_count(line):
    if 'fantastic' in line:
        return line['fantastic']
    else:
        return 0
products['fantastic'] = products['word_count'].apply(fantastic_count)
def amazing_count(line):
    if 'amazing' in line:
        return line['amazing']
    else:
        return 0
products['amazing'] = products['word_count'].apply(amazing_count)
def love_count(line):
    if 'love' in line:
        return line['love']
    else:
        return 0
products['love'] = products['word_count'].apply(love_count)
def horrible_count(line):
    if 'horrible' in line:
        return line['horrible']
    else:
        return 0
products['horrible'] = products['word_count'].apply(horrible_count)
def bad_count(line):
    if 'bad' in line:
        return line['bad']
    else:
        return 0
products['bad'] = products['word_count'].apply(bad_count)
def terrible_count(line):
    if 'terrible' in line:
        return line['terrible']
    else:
        return 0
products['terrible'] = products['word_count'].apply(terrible_count)
def awful_count(line):
    if 'awful' in line:
        return line['awful']
    else:
        return 0
products['awful'] = products['word_count'].apply(awful_count)
def wow_count(line):
    if 'wow' in line:
        return line['wow']
    else:
        return 0
products['wow'] = products['word_count'].apply(wow_count)
def hate_count(line):
    if 'hate' in line:
        return line['hate']
    else:
        return 0
products['hate'] = products['word_count'].apply(hate_count)
print sum(products['awesome'])
print sum(products['great'])
print sum(products['fantastic'])
print sum(products['amazing'])
print sum(products['love'])
print sum(products['horrible'])
print sum(products['bad'])
print sum(products['terrible'])
print sum(products['awful'])
print sum(products['wow'])
print sum(products['hate'])
```

#build a sentiment classifier

```python
products['rating'].show(view='Categorical')
```

##define negative and postive

```python
#ignore all 3* reviews
products=products[products['rating']!=3]
```

```python
#positive sentiment =4* or 5*
products['sentiment']=products['rating']>=4
```

```python
products.head()
```

#let's train the sentiment classifier

```python
train_data,test_data=products.random_split(.8,seed=0)
```

```python
select_words_model=graphlab.logistic_classifier.create(train_data,
                                                    target='sentiment',
                                                    features=selected_words,
                                                    validation_set=test_data)
```

#evaluate the sentiment model

```python
select_words_model.evaluate(test_data,metric='roc_curve')
```

```python
select_words_model.show(view='Evaluation')
```

```python
select_words_model['coefficients']
```
